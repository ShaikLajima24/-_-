```python
#CAR PRICE PREDICTION PROJECT

#IMPORT NECESSARY LIBRARIES

from google.colab import files
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer # Corrected spelling
from sklearn.preprocessing import OneHotEncoder # Corrected spelling and module
from sklearn.metrics import mean_squared_error,r2_score
```

```python

# UPLOAD CSV FILE
uploaded=files.upload()
```

```python

# LOAD DATA SET
df=pd.read_csv(next(iter(uploaded)))
```

```python

# DISPLAY THE DATASET
df.head()
```

```python
#SPLIT DATASET INTO FEATURES AND TARGET
x=df[['Year','Present_Price','Driven_kms','Fuel_Type','Selling_type','Owner','Transmission']]
y=df[['Selling_Price']]
```

```python

#ENCODE CATEGORICAL VARIABLES
categorical_columns = ['Year','Driven_kms','Fuel_Type','Selling_type','Owner','Transmission']
# Corrected the spelling of ColumnTransformer and categorical_coluumns
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(drop='first', handle_unknown='ignore'), categorical_columns)], remainder='passthrough')
# Corrected the spelling of fit_transformer
x_encoded = ct.fit_transform(x)
```

```python
# Split into train and test sets
x_train, x_test, y_train, y_test = train_test_split(x_encoded, y, test_size=0.2, random_state=42)

# Train Linear Regression model
model = LinearRegression()
model.fit(x_train, y_train)

# Predict
y_pred = model.predict(x_test)
```

```python
# Evaluate the model
print(f"Predicted car selling price is: {y_pred[0][0]:.2f}")
print(f"Actual car prices:\n{y_test.values}")
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
print("R² Score:", r2_score(y_test, y_pred))
```

```python
# Predict on new input (no `input()` in Colab, so use variables or forms)
# You can modify these values to test different cases
yearinput = 2017
presentpriceinput = 9.85
drivenkmsinput = 6900
fueltypeinput = 'Petrol'
sellingtypeinput = 'Dealer'
ownerinput = 0
transmissioninput = 'Manual'

# Create input DataFrame
newcar_df = pd.DataFrame([{
    'Year': yearinput,
    'Present_Price': presentpriceinput,
    'Driven_kms': drivenkmsinput,
    'Fuel_Type': fueltypeinput,
    'Selling_type': sellingtypeinput,
    'Owner': ownerinput,
    'Transmission': transmissioninput
}])

# Encode and predict
newcar_encoded = ct.transform(newcar_df)
predicted_price = model.predict(newcar_encoded)

print(f"Predicted Selling Price: ₹ {predicted_price[0][0]:,.2f}")
```
